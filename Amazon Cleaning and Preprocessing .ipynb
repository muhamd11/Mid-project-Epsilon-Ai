## Preprocessing Data

```python
df.head(5)
```

### Delete unnecessary columns

```python
df.drop(['user_id','user_name','review_id','review_title','review_content','img_link','img_link','product_link','about_product'],axis=1,inplace=True)
```

```python
df.head(5)
```

### Convert category column into main category column & sub category column 

```python
cat = df['category'].str.split('|',expand=True)
```

```python
cat.rename(columns={0:'main category',1:'sub_cat1',2:'sub_cat2',3:'sub_cat3',4:'sub_cat4'},inplace=True)
```

```python
cat
```

```python
df['main category'] = cat['main category']
df['sub category']  = cat['sub_cat3']
df.drop('category',axis=1,inplace=True)
```

### cleaning and change data type of discounted_price column & discounted_price column 

```python
df['discounted_price'] = df['discounted_price'].str.replace('₹','')
df['actual_price'] = df['actual_price'].str.replace('₹','')
df['discounted_price'] = df['discounted_price'].str.replace(',','')
df['actual_price'] = df['actual_price'].str.replace(',','')
df['discount_percentage %'] = df['discount_percentage'].str.replace('%','')
```

```python
df['discounted_price'] = df['discounted_price'].astype('float64')
df['actual_price'] = df['actual_price'].astype('float64')
df['discount_percentage %'] = df['discount_percentage %'].astype('float64')
df.drop('discount_percentage',axis=1,inplace=True)
```

```python
df.info()
```

```python
df.head(5)
```

### cleaning and change data type of rating & rating columns

```python
df[df['rating'] == '|']
df.drop(index=1279,axis=0,inplace=True)
```

```python
df[df['rating'] == '|']
```

```python
df['rating'] = df['rating'].astype('float64')
```

```python
df['rating_count'] = df['rating_count'].str.replace(',','')
df['rating_count'] = df['rating_count'].astype('float64')
```

```python
df.head(5)
```

```python
rating_scale = []
for i in df['rating']:
    if i < 2:
        rating_scale.append('bad')
    elif i < 3:
        rating_scale.append('good')    
    elif i < 4:
        rating_scale.append('very good')    
    elif i >= 4:
        rating_scale.append('excellent')    
        
```

```python
df['quality of product'] = rating_scale
```

### Filling and deleting missing values 

```python
df.isna().sum()
```

```python
from sklearn.impute import KNNImputer
```

```python
imputer = KNNImputer()
```

```python
df['rating_count'] = imputer.fit_transform(df[['rating_count']])
```

```python
df.isna().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.isna().sum()
```

### Handling duplicated values

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.duplicated().sum()
```
